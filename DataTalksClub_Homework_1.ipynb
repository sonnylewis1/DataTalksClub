```python
companies = ['2222.SR','BRK-B', 'AAPL', 'MSFT', 'GOOG', 'JPM']
stock_ohlcv = {}

for company in companies:
  comp_daily_non_delayed = yf.download(tickers = company,
                     period = "max",
                     interval = "1d", start = date(year=2023, month=1, day=1), end = date(year=2023, month=12, day=31))
  ratio = (max(comp_daily_non_delayed['Adj Close']) - min(comp_daily_non_delayed['Adj Close'])) / max(comp_daily_non_delayed['Adj Close'])
  stock_ohlcv[company] = ratio

stock_ohlcv
```

**QUESTION 5**

```python
companies = ['2222.SR','BRK-B', 'AAPL', 'MSFT', 'GOOG', 'JPM']
yeild_all = {}

for company in companies:
  comp_stock = yf.Ticker(company)

  comp_dividends = sum(comp_stock.dividends['2023-01-01' : '2023-12-31'])

  comp_daily_non_delayed = yf.download(tickers = company,
                     period = "max",
                     interval = "1d", start = date(year=2023, month=1, day=1), end = date(year=2023, month=12, day=31))

  dividend_yeild = comp_dividends / comp_daily_non_delayed['Adj Close'][-1]
  yeild_all[company] = dividend_yeild * 100
```

```python
dict(sorted(yeild_all.items(), key=lambda item: item[1]))
```
